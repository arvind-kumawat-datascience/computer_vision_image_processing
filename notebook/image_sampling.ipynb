#### Image Sampling and Quantization

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Read the high-resolution image
image = cv2.imread('../img/sample.jpeg')  # Replace with your image path
```

```python
# Step 2: Downsample the image (reduce resolution)
downsampled_image = cv2.resize(image, (image.shape[1]//2, image.shape[0]//2))
```

```python
# Step 3: Apply quantization (8-bit, 4-bit, and 2-bit)
quantization_levels = [256, 16, 4]
```

```python
quantized_images = []
for levels in quantization_levels:
    quantized_image = np.floor(downsampled_image / (256 // levels)) * (256 // levels)
    quantized_images.append(quantized_image)
```

```python
# Step 4: Display the results
fig, axes = plt.subplots(1, 4, figsize=(15, 5))

axes[0].imshow(downsampled_image)
axes[0].set_title("Downsampled Image")
axes[1].imshow(quantized_images[0])
axes[1].set_title("8-bit Quantization")
axes[2].imshow(quantized_images[1])
axes[2].set_title("4-bit Quantization")
axes[3].imshow(quantized_images[2])
axes[3].set_title("2-bit Quantization")
plt.show()
```
