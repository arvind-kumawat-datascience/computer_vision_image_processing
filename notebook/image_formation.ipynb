#### Image Formation

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
```

```python
# Step 1: Create a synthetic sinusoidal image
x = np.linspace(0, 2 * np.pi, 256)
y = np.linspace(0, 2 * np.pi, 256)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) + np.cos(Y)
```

```python
# Step 2: Display the generated sinusoidal pattern
plt.imshow(Z, cmap='gray')
plt.title("Synthetic Sinusoidal Image")
plt.show()
```

```python
# Step 3: Save the image in multiple formats
cv2.imwrite('../img/imf1.bmp', Z * 255)  # BMP format
cv2.imwrite('../img/imf2.png', Z * 255)  # PNG format
cv2.imwrite('../img/imf3.jpg', Z * 255)  # JPEG format
```

```python
# Step 4: Check file sizes (optional step)
bmp_size = os.path.getsize('../img/imf1.bmp')
png_size = os.path.getsize('../img/imf2.png')
jpg_size = os.path.getsize('../img/imf3.jpg')
```

```python
print(f"BMP Size: {bmp_size} bytes")
print(f"PNG Size: {png_size} bytes")
print(f"JPEG Size: {jpg_size} bytes")
```
