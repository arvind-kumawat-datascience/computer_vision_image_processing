#### Pre-processing and Binary Image Analysis

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Read the grayscale image
image = cv2.imread('../img/sample.jpeg', cv2.IMREAD_GRAYSCALE)
```

```python
# Step 2: Apply global thresholding (Otsu's method)
_, binary_image = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
# Step 3: Apply morphological operations (dilation and erosion)
kernel = np.ones((5, 5), np.uint8)
dilated_image = cv2.dilate(binary_image, kernel, iterations=1)
eroded_image = cv2.erode(binary_image, kernel, iterations=1)
```

```python
# Step 4: Perform connected component labeling
num_labels, labels = cv2.connectedComponents(binary_image)
```

```python
# Step 5: Display the results
fig, axes = plt.subplots(1, 4, figsize=(15, 5))
axes[0].imshow(binary_image, cmap='gray')
axes[0].set_title("Binary Image")
axes[1].imshow(dilated_image, cmap='gray')
axes[1].set_title("Dilated Image")
axes[2].imshow(eroded_image, cmap='gray')
axes[2].set_title("Eroded Image")
axes[3].imshow(labels, cmap='jet')
axes[3].set_title(f"Connected Components: {num_labels} labels")
plt.show()
```
