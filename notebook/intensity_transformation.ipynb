#### Intensity Transformation and Analysis

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Read a low-contrast image
image = cv2.imread('../img/sample.jpeg', cv2.IMREAD_GRAYSCALE)
```

```python
# Step 2: Apply contrast stretching
min_pixel = np.min(image)
max_pixel = np.max(image)
contrast_stretched = ((image - min_pixel) / (max_pixel - min_pixel)) * 255
```

```python
# Step 3: Apply histogram equalization
hist_eq_image = cv2.equalizeHist(image)
```

```python
# Step 4: Display original, contrast stretched, and histogram equalized images
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(image, cmap='gray')
axes[0].set_title("Original Image")
axes[1].imshow(contrast_stretched, cmap='gray')
axes[1].set_title("Contrast Stretched")
axes[2].imshow(hist_eq_image, cmap='gray')
axes[2].set_title("Histogram Equalized")
plt.show()
```

```python
# Step 5: Plot histograms
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(image.ravel(), bins=256, range=(0, 255))
plt.title("Histogram of Original Image")
plt.subplot(1, 2, 2)
plt.hist(hist_eq_image.ravel(), bins=256, range=(0, 255))
plt.title("Histogram of Equalized Image")
plt.show()
```
